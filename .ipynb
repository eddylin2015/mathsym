```python
"""
Get_OneVarTwoExpEquExpr
十字相乘法因式分解factor   x**2 + b*x + c
"""
####################初始化,文件頭,匯入組件#####################################
import random                                     #亂數 
import math                                       #math 內置數學函數
import numpy as np                                #數字矩陣
import sympy as sp                                #sympy 簡易別名 sp    
from sympy.parsing.sympy_parser import parse_expr #文字字串, 解釋成, Sympy 運算式
from sympy.plotting import plot                   #繪圖表
from IPython.display import Latex,HTML                 #網頁顯示數學符號
import json                                       #JSON 結構化資料
sp.init_printing("mathjax")                       #sp.init_printing()  168 
############################################################################
def Get_OneVarTwoExpEquExpr(QizAmt):
    sample_list0= list(range(-39,29))   # [-5,-4,-3,-2,-1,1,2,3,4,5]
    sample_list1= list(range(-39,29))
    sample_list1.remove(0)    # 非零數列
    NTE=[]
    for i in range(0,QizAmt):
        a = 1 
        p=random.choice([-5,-4,-3,-2,-1,1,2,3,4,5])
        q=random.choice([-5,-4,-3,-2,-1,1,2,3,4,5])
        b=p+q
        c=p*q
        express_str=f"{a}*x**2 + {b}*x +  {c} "  
        St=parse_expr(express_str, evaluate=False)     #字串解釋為可運算式子 expression
        Val=sp.factor(St)                              #因式分解,得出標準答案
        TE={}
        TE["St"]=sp.latex(St)
        TE["Val"]=Val        
        NTE.append(TE)    
    return NTE
QizAmt=4
Mark=0
NTE=Get_OneVarTwoExpEquExpr(QizAmt)
i=0
for TE in NTE:
    i+=1 
    St=TE["St"]
    Val=TE["Val"]
    print(f"第{i}題:")        
    display(Latex(f" $$ {St} $$ " ))
    ans=input("請作答:")  
    TE["Ans"]=ans
    if parse_expr(ans)==Val:                   #比對答案:
        print("答對!")                              #答對加一分
        Mark+=1
        TE["OK"]=1
    else:                                      #不則
        TE["OK"]=0
        display(Latex(f"答案錯誤, 標準答案: $ {Val} $"))
    

print(f"總結: 共{QizAmt}題,得 {Mark} 分. ")  #總結得分

tablecontext="<table><tr><td>Qiz</td><td>Val</td><td>Ans</td><td>OK</td></tr>"
for row in NTE:
    St=row["St"]
    Val=row["Val"]
    Ans=row["Ans"]
    OK=row["OK"]
    tablecontext+=f"<tr><td>$${St}$$</td><td>{Val}</td><td>{Ans}</td><td>{OK}</td></tr>"
tablecontext+="</table>"
display(HTML(tablecontext))    
```
