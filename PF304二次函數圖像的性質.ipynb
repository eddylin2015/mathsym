二次函數圖像的性質   

題型1	答案式样	数型及范围  
y=a(x-h)**2 + k  
頂點坐標是(h,k)  .	答: (h,k).	  
K為非0整數  
h,k為整數  

```python
import random                                     #亂數 
import math                                       #math 內置數學函數
import numpy as np                                #數字矩陣
import sympy as sp                                #sympy 簡易別名 sp    
from sympy.parsing.sympy_parser import parse_expr #文字字串, 解釋成, Sympy 運算式
from sympy.plotting import plot                   #繪圖表
from IPython.display import Latex,HTML            #網頁顯示數學符號
import json                                       #JSON 結構化資料
import re
import lib
sp.init_printing("mathjax")                       
#sp.init_printing()  
def displayTable(NTE):
    tablecontext="<table><tr><td>{}</td></tr>".format("</td><td>".join(_ for _ in ["Qiz","Val","Ans","OK"]))
    for row in NTE:
        St=row["St"]
        Val=row["Val"]
        Ans=row["Ans"]
        OK=row["OK"]
        tablecontext+=f"<tr><td>$${St}$$</td><td>{Val}</td><td>{Ans}</td><td>{OK}</td></tr>"
    tablecontext+="</table>"
    display(HTML(tablecontext)) 
"""
二次函數圖像的性質
"""
def Get_PF304_Expr(QizAmt):
    x,y,z=sp.symbols('x,y,z')
    sample_list0= list(range(-10,10))   # [-5,-4,-3,-2,-1,1,2,3,4,5]
    sample_list1= list(range(-10,10))
    sample_list1.remove(0)    # 非零數列
    NTE=[]
    for i in range(0,QizAmt):
        a = random.choice(sample_list1) 
        h =random.choice(sample_list0)
        k =random.choice(sample_list0)
        St=sp.Eq(y, a*(x-h)**2+k )  #, evaluate=False) #字串解釋為可運算式子 expression
        Val= [h,k]
        TE={}
        TE["Plot"]=a*(x-h)**2+k 
        TE["St"]=sp.latex(St)
        TE["Val"]=Val        
        TE["Ans"]=""
        TE["OK"]=0
        TE["Tip"]=""
        NTE.append(TE)    
    return NTE

if __name__=="__main__":
    NTE=Get_PF304_Expr(10)
    displayTable(NTE)   
```

```python
x,y,z=sp.symbols('x,y,z')
NTE=Get_PF304_Expr(4)
for i,TE in enumerate(NTE):
    St=TE["St"]
    Val=TE["Val"]
    print(f"第{i}題: 頂點坐標是____  ? ")        
    display(Latex(f" $$ {St} $$ " ))
    plot(TE["Plot"],(x,-100,100))
    print(TE["Tip"])
    #print(Val)
    ans=[]
    for idx in range(0,len(Val)):
        ax_title=["x","y"]
        ax=ax_title[idx % 2]
        ans1=input(f"請作答 {ax}:") 
        if ans1=="" : ans1="0"
        ans.append(parse_expr(lib.Text2St(ans1)))
    TE["Ans"]=ans
    Flag=0
    for v_ in Val:
        for a_ in ans:
            if(v_==a_): Flag+=1
    if Flag>1:                   #比對答案:
        print("答對!")                              #答對加一分
        TE["OK"]=1
    else:                                           #不則
        TE["OK"]=0
        display(Latex(f"答案錯誤, 標準答案: $ {Val} $"))

displayTable(NTE)     
```

```python
"""
二次函數圖像的性質

題型1	答案式样	数型及范围
y=a(x-h)**2 + k
頂點坐標是(h,k)  .	答: (h,k).	
K為非0整數
h,k為整數

"""
####################初始化,文件頭,匯入組件#####################################
import random                                     #亂數 
import sympy as sp                                #sympy 簡易別名 sp    
from sympy.parsing.sympy_parser import parse_expr #文字字串, 解釋成, Sympy 運算式
from sympy.plotting import plot                   #繪圖表
from IPython.display import Latex,HTML,Markdown            #網頁顯示數學符號
import lib
sp.init_printing("mathjax")                       
#sp.init_printing()  
def displayTable(NTE):
    tablecontext=lib.NTE2MD(NTE)
    display(Markdown(tablecontext)) 
############################################################################
def Get_RelationalNumberExpr(QizAmt):
    x,y,z=sp.symbols('x,y,z')
    sample_list0= list(range(-10,10))   # [-5,-4,-3,-2,-1,1,2,3,4,5]
    sample_list1= list(range(-10,10))
    sample_list1.remove(0)    # 非零數列
    NTE=[]
    for i in range(0,QizAmt):
        a = random.choice(sample_list1) 
        h =random.choice(sample_list0)
        k =random.choice(sample_list0)
        
        St=sp.Eq(y,a*(x-h)**2+k )  #, evaluate=False) #字串解釋為可運算式子 expression
        Val= [h,k]
        
        TE={}
        TE["St"]=sp.latex(St)
        TE["Val"]=Val        
        TE["Ans"]="-"
        TE["OK"]=0
        TE["Plot"]=a*(x-h)**2+k 
        NTE.append(TE)    
    return NTE

x,y,z=sp.symbols('x,y,z')
QizAmt=4
NTE=Get_RelationalNumberExpr(QizAmt)
for TE in enumerate(NTE):
    St=TE["St"]
    Val=TE["Val"]
    print(f"第{i+1}題: 頂點坐標是____  ? ")        
    display(Latex(f" $$ {St} $$ " ))
    plot(TE["Plot"],(x,-100,100))
    display(Latex(f"答案錯誤, 標準答案: $ {Val} $"))

print(f"總結: 共{QizAmt}題,得 {Mark} 分. ")  #總結得分

displayTable(NTE)     
```
