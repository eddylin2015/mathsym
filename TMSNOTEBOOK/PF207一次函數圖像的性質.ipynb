題目: y隨x的增大而_____.	(增大 +1 或  減少 -1). k , b 有是整數.

$$y=kx+b$$

作答: +1 或 -1


```
一次函數圖像的性質
題型1	答案式样	数型及范围
y=kx+b
y隨x的增大而_____.	答: 增大  或  減少.	
K為非0整數
b為整數

命题方式参考	
K隨機取±10以內的非0整數, b隨機取±10以內的整數;
k>0 時 , 答案為”增大”;
k<0 時 , 答案為”增大”.

題型2	答案式样	数型及范围
y=kx+b圖像經過點(m,____) 或 (____, n)
答: ( x , y )	
K為非0實數
B為實數

命题方式参考	
K隨機取±10以內的非0實數, b隨機取±10以內的實數;
則x=m , y=km+b或 x=(y-b)/k , y=n

題型3	答案式样	数型及范围
y=kx+b
x=m 時y=n , 
又 x=p時y=q
求k 與q.	答: k=____ , b=_____	K為非0實數
b為實數

命题方式参考	
K隨機取±10以內的非0實數, b隨機取±10以內的實數;
k=((n-q))/((m-p)) , b=n-(m(n-q))/((m-p)) .

題型4	答案式样	
数型及范围
已知y+p與x+q成正比例,且x=m時y=n,求函數的關係式.	答:y=kx+b	p,q,m,n隨機取±10以內的整數
命题方式参考	p,q,m,n隨機取±10以內的整數
k=((n+p))/((m+q)) , b=n
```

````python
import random                                     #亂數 
import math                                       #math 內置數學函數
import numpy as np                                #數字矩陣
import sympy as sp                                #sympy 簡易別名 sp    
from sympy.parsing.sympy_parser import parse_expr #文字字串, 解釋成, Sympy 運算式
from sympy.plotting import plot                   #繪圖表
from IPython.display import Latex,HTML,Markdown            #網頁顯示數學符號
import json                                       #JSON 結構化資料
import re
import lib
from lib import GetTE
#sp.init_printing("mathjax")                       
#sp.init_printing()  
"""
一次函數圖像的性質
"""
def Get_PF207_Expr(QizAmt):
    x,y,z=sp.symbols('x,y,z')
    sample_list0= list(range(-10,10))   # [-5,-4,-3,-2,-1,1,2,3,4,5]
    sample_list1= list(range(-10,10))
    sample_list1.remove(0)    # 非零數列
    NTE=[]
    for i in range(0,QizAmt):
        k = random.choice(sample_list1) 
        b =random.choice(sample_list0)
        St=sp.Eq(y, k*x+b )  #, evaluate=False) #字串解釋為可運算式子 expression
        Val= 1
        if k<0 : Val=-1
        TE=GetTE(len(NTE),sp.latex(St),Val)
        TE["Plot"]=k*x+b
        TE["Tip"]="y隨x的增大而____  (  +1 表示 增大  或  -1 表示  減少 )"
        NTE.append(TE)    
        
    return NTE

NTE=Get_PF207_Expr(10)
display(Markdown(lib.NTE2TBL(NTE)))
````

````python
x,y,z=sp.symbols('x,y,z')
QizAmt=4
Mark=0
NTE=Get_PF207_Expr(QizAmt)
i=0
for TE in NTE:
    i+=1 
    St=TE["St"]
    Val=TE["Val"]
    print(f"第{i}題: y隨x的增大而____  (  +1 表示 增大  或  -1 表示  減少 ) ")        
    display(Latex(f" $$ {St} $$ " ))
    plot(TE["Plot"])
    ans=input("請作答:")  
    TE["Ans"]=ans
    if int(ans)==Val:                   #比對答案:
        print("答對!")                              #答對加一分
        Mark+=1
        TE["OK"]=1
    else:                                           #不則
        TE["OK"]=0
        display(Latex(f"答案錯誤, 標準答案: $ {Val} $"))

print(f"總結: 共{QizAmt}題,得 {Mark} 分. ")  #總結得分
display(Markdown(lib.NTE2TBL(NTE)))

````
