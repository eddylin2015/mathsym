# 同角三角函数的关系

同角：终边相同的角

三角函数：正弦、余弦、正切函数

关系：正弦、余弦的平方关系；正弦、余弦的商数关系

总结：

同一个角的正弦、余弦的平方和等于1，商等于这个教的正切。

## 代数式表示

### 平方关系：
sin²α+cos²α=1。   

### 商数关系：

sinα/cosα=tanα=secα/cscα、cosα/sinα=cotα=cscα/secα；  

$\tan {\theta}=\frac{\sin{\theta}}{\cos{\theta}} = \frac{\sec{\theta}}{\csc{\theta}} $

$\cot{\theta}=\frac{\cos{\theta}}{\sin{\theta}} = \frac{\csc{\theta}}{\sec{\theta}} $

## 同角三角函数的基本关系式  



倒数关系：tanα ·cotα=1、sinα ·cscα=1、cosα ·secα=1；    
商的关系： sinα/cosα=tanα=secα/cscα、cosα/sinα=cotα=cscα/secα；   
和的关系：sin2α+cos2α=1、1+tan2α=sec2α、1+cot2α=csc2α；   
平方关系：sin²α+cos²α=1。
