$$4x^1*y^3+x^1*(−5)y^2$$

第3題:
$$6x^3*y^2+x^3*(−8)*y^1$$

第4題:
$$ 3*x^1*y^3+3*x^0*y^5 $$
電腦標準答案: 3*y**3*(x + y**2)


因式分解 提公因式  
題型1：  
"{ab} * x**{p} * y**{q}  + {ac} * x**{u}*y**{v}  "   
s1=abxpyq, s2=acxuyv, s3=adxgyh, s1’= bxp-myq-n, s2’= cxu-myv-n, s3’= dxg-myh-n,   
s1 +s2 +s3=axmyn(s1’+ s2’+ s3’) (s1≠s2≠s3), s1 ,s2 ,s3亂序,s和s’的順序相同  
(|a|,|b|,|c|,|d|,|k|)<13 整數, -1<(p,q,u,v,g,h)<11整數, (a,b,c)≠0, m=min(p,u,g), n=min(q,v,h),   
(x,y)=(a至z內隨機字母) 或 (a至z內隨機字母 ± a至z內隨機字母) 或 (a至z內隨機字母 ± k), x≠y  

```python
import math                                       #math 內置數學函數
import numpy as np                                #數字矩陣
import re
import random                                     #亂數 
import sympy as sp                                #sympy 簡易別名 sp    
from sympy.parsing.sympy_parser import parse_expr #文字字串, 解釋成, Sympy 運算式
from IPython.display import Latex,HTML,Markdown                 #網頁顯示數學符號
import lib
from lib import GetTE
sp.init_printing("mathjax")                       #sp.init_printing()  168 
"""
因式分解 提公因式
"""
def Get_PF203_Expr(QizAmt):
    NTE=[]
    for i in range(0,QizAmt):
        a =  random.choice([-5,-4,-3,-2,-1,1,2,3,4,5])
        b =  random.choice([-5,-4,-3,-2,-1,1,2,3,4,5])
        c =  random.choice([-5,-4,-3,-2,-1,1,2,3,4,5])
        d =  random.choice([-5,-4,-3,-2,-1,1,2,3,4,5])
        ab=a*b
        ac=a*c
        ad=a*d
        p=random.choice([0,1,2,3,4,5])
        q=random.choice([0,1,2,3,4,5])
        u=random.choice([0,1,2,3,4,5])
        v=random.choice([0,1,2,3,4,5])
        
        express_str=f"{ab} * x**{p} * y**{q}  + {ac} * x**{u}*y**{v}  "  #題型 express_str ax+bx+c
        St=parse_expr(express_str, evaluate=False) #字串解釋為可運算式子 expression
        Val=sp.factor(St)
        TE=GetTE(len(NTE),sp.latex(St), Val)
        TE["ValLatex"]=sp.latex(Val)    
        NTE.append(TE)    
    return NTE

x,y,z=sp.symbols('x,y,z')
NTE=Get_PF203_Expr(4)
display(Markdown(lib.NTE2TBL(NTE)))  
```

```python
"""程式二
因式分解 提公因式
"""
x,y,z=sp.symbols('x,y,z')
QizAmt=4
Mark=0
NTE=Get_PF203_Expr(QizAmt)
for i, TE in enumerate(NTE):
    St=TE["St"]
    Val=TE["Val"]
    ValLatex=TE["ValLatex"]
    print(f"第{i}題:")        
    display(Latex(f" $$ {St} $$ " ))
    #display(Latex(f"電腦標準答案: $ {ValLatex} $ ; {Val}"))
    ans=input("請作答:")  
    ans=lib.Text2St(ans)
    TE["Ans"]=ans
    if parse_expr(ans).subs({x:7})==Val.subs({x:7}):                   #比對答案:
        print("答對!")                              #答對加一分
        Mark+=1
        TE["OK"]=1
    else:                                      #不則
        TE["OK"]=0
        display(Latex(f"答案錯誤, 標準答案: $ {Val} $"))

print(f"總結: 共{QizAmt}題,得 {Mark} 分. ")  #總結得分

display(Markdown(lib.NTE2TBL(NTE)))  
```
