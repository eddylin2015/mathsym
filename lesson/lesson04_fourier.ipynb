### 7.ÁØÑ‰æã11-2ÔºöÊØîËºÉ‰∏çÂêånÁ¥öÊï∏ÁöÑÂÇÖÁ´ãËëâÁ¥öÊï∏Fourier SeriesÂéªÊ®°Êì¨‰∏âÊ¨°ÊñπÂ§öÈ†ÖÂºèÁöÑÁ∑öÂûãÊïàÊûú

```python

#Ôºà1ÔºâËã•ÊòØnÁ¥öÊï∏Â§™Â∞ë(n=5)ÔºåÂ∞±ÊúÉÂΩ¢ÊàêÈã∏ÈΩíÁãÄÊ®°Êì¨ÁµêÊûú
#Ôºà2ÔºâËã•ÊòØnÁ¥öÊï∏Â§†Â§ß(n=40))ÔºåÂ∞±ÊúÉÂΩ¢ÊàêÂπ≥ÊªëÂåñÁöÑÊ®°Êì¨ÁµêÊûú
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

x = symbols('x')
f = Function('f')(x)
f = x**3 + x**2 + x +1
#Ë™ûÊ≥ïÔºöfourier_series(Ë¶ÅÊ®°Êì¨ÁöÑfÂáΩÊï∏, (xÁöÑÁØÑÂúç, -pi, pi))
s = fourier_series(f, (x, -pi, pi))
#‰∏çÂ§™ÂèØËÉΩÂéª‰ΩøÁî®n=‚àûÂéªË®àÁÆóÔºåÂú®ÂØ¶ÈöõË®àÁÆóÊôÇË¶ÅÊà™Áü≠Á¥öÊï∏truncateÔºàÊåáÂÆön=Êï∏Â≠óÔºâ
fs5 = s.truncate(n=5)
fs10 = s.truncate(n=40)

#Áï´Âúñ
p1 = plot(fs5,fs10,show=False)
p1[0].line_color = 'red'
p1[1].line_color = 'green'
p1[0].label = '$fourier(n=5)$'
p1[1].label = '$fourier(n=40)$'
p1.legend = True
p1.show()
```
