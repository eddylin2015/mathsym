費馬大定理 英國著名數學家安德魯 · 懷爾斯

    由十七世紀法國數學家費馬提出的“費馬大定理”：“設n是大於二的正整數，則方程式$x^n+y^n=z^n$，沒有正整數解”，相信大家都不會陌生。一六三七年，有“業餘數學家”封號的費馬，在閱讀丟番圖的《算術》拉丁文譯本時，他就在書上寫上了一道算式，並拋下了這句話。就這樣，費馬這一個隨便的猜想，譜寫了一段精彩絕倫的“數學史”。

    “費馬大定理”前後共走過了三個世紀，經過一代又一代傑出數學家的努力，最終由英國著名數學家安德魯 · 懷爾斯完成了對“費馬大定理”的證明。懷爾斯教授在成功之前也可算是困難重重，過程更是曲折離奇。當年，他正準備透過研討會向世界公佈研究結果，在進入最後審查階段的關鍵時刻，才找到一處先前完全沒有發現的錯誤。就在陷入絕望之際，他在辦公室桌上看到一份“岩澤理論”的論文，就隨手運用了岩澤理論來化解原先證明的錯誤。終於，在一九九五年，足足等待了三百年的世紀難題正式解開了。
