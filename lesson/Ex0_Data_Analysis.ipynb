```python
fns="""右腹股溝韌帶邊緣內側(無重叠)	A內下	A外下	A上下	Ast50pr	Agt50pr	A1	左腹股溝韌帶邊緣左內側(無重叠)	B內下	B外下	B上下	Bst50pr	Bgt50pr	B1	B雙側	B重叠率*	右韌帶下方2CM內側	C內下	C外下	C上下	Cst50pr	Cgt50pr	C1	左韌帶下方2CM內側	D內下	D外下	D上下	Dst50pr	Dgt50pr	D1	D雙側	D重叠率*"""
fns=fns.split("\t")
for fn in fns:
    df1 = df[df[fn]=="X"]
    mean=round(df1["年齡"].mean(),2)
    std=round(df1["年齡"].std(),2)
    cnt=df1.shape[0]
    print(f"{fn}  有{cnt}筆年齡: {mean}+-{std}")
    M_F=df1["性別"].value_counts()
    M=M_F["M"]
    Mr=round(M/cnt,4)
    F=M_F["F"]
    Fr=round(F/cnt,4)
    print(f"{M} {Mr}  {F} {Fr}")
    displaySer(M_F)
    
#print(df1.shape[0])
#df1=  df1=="X" 
#sum_s=df1.sum(axis=0)
#display(sum_s)
#displaySer(sum_s)
```
