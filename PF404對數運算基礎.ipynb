### 簡化包含三角函數的運算式
使用 簡化包含三角函數和雙曲函數的運算式。trigsimp

$cos^2 x−sin^2 x $  "簡化"。

```python
x=sp.Symbol('x')
sp.trigsimp(sp.cos(x)**2 - sp.sin(x)**2)
```

$cosh^2x−sinh^2x$   "簡化"。

```python
sp.trigsimp(sp.cosh(x)**2 - sp.sinh(x)**2)
```

## 第5節: 概率及亂數

Let's roll two dice, X and Y, with six faces each

```python
import sympy as sp 
from sympy.stats import *
sp.init_printing()
#Let's roll two dice, X and Y, with six faces each
X, Y = Die('X', 6), Die('Y', 6)
```

```python
P(sp.Eq(X, 3))
```

```python
P(X > Y)
```

```python
P(X + Y > 6, X < 5)
```
