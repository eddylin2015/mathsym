# 等比數列之和

```python
1-2+4-8+ 16 -32 +64- 128+ 256 -512 + 1024 
```

```python
exper =-2**(-1)
```

```python
import sympy as sp
import numpy as np
n,i= sp.symbols('n,i')
expre= (-2) ** (n-1)
expr_func= sp.lambdify(n, expre)
n_arr=np.arange(12)
n_arr[0]=1
a=expr_func(n_arr)
a
```

### 等比

```python

q=a[4]/a[3]
q
```

### 通項

```python

expre.subs(n,11)
```

```python
a[1] * q**(11-1)
```

### 和

```python
(a[1]*(1-q**n)/(1-q)).subs(n,11) 
```

```python
(a[1]-a[11]*q)/(1-q)
```

```python
sp.summation( expre, (n, 1, 11))     
```
