```python
import random                                     #亂數 
import sympy as sp                                #sympy 簡易別名 sp    
from sympy.parsing.sympy_parser import parse_expr #文字字串,解釋成運算式
from IPython.display import Latex,HTML                 #網頁顯示數學符號
sp.init_printing("mathjax")                       #sp.init_printing()  168 
import lib
def displayTable(NTE):
    display(HTML(lib.NTE2HTMLTABLE(NTE))) 
############################################################################
"""
根式練習
"""
def Put_PF201_Expr(TE):
    x=sp.Symbol('x')
    Val=TE["Val"]
    ans=lib.Text2St(TE["Ans"])
    try:
        if parse_expr(ans)==Val:                   #比對答案:
            TE["OK"]=1
        else:                                      #不則
            TE["OK"]=0
    except:
        pass
    return TE["OK"]==1

def Post_PF201_Expr(NTE):
    x=sp.Symbol('x')
    for TE in NTE:
        Put_PF201_Expr(TE)

def Get_PF201_Expr(QizAmt):
    sample_list0= list(range(-39,29))   # [-5,-4,-3,-2,-1,1,2,3,4,5]
    sample_list1= list(range(8,20))    #sample_list1.remove(0)    # 非零數列
    b_li=list(range(1,18))
    b_li.remove(4)
    b_li.remove(9)
    b_li.remove(16)
    
    NTE=[]
    for i in range(0,QizAmt):
        Tx=i%4
        if Tx==1:
            a=random.choice(range(2,18))
            b=random.choice(b_li)
            c=random.choice(range(2,15))
            d=random.choice(range(2,15))
            m=c*c*b
            n=d*d*b
            St=r"\sqrt{%s} + \sqrt{%s} "% (m,n)
            Val=sp.sqrt(m)+sp.sqrt(n)
        elif Tx==2:
            a=random.choice(range(1,11))
            b=random.choice(b_li)
            c=random.choice(range(1,10))
            d=random.choice(range(1,10))
            e=random.choice([-1,-1,1])
            m=c*c*b
            n=d*d*b
            if e==-1:
                St=r"\sqrt{%s} \times {(\sqrt{%s})}^{%s} "% (m,n,e)
            else:
                St=r"\sqrt{%s} \times {\sqrt{%s}} "% (m,n)
            Val=sp.sqrt(m)*sp.sqrt(n)**e
        elif Tx==3:
            a=random.choice(range(1,11))
            c=random.choice(range(1,10))
            b=random.choice(b_li)
            d=random.choice(b_li)
            l=random.choice(range(1,10))
            e=random.choice([-1,1])
            m=a*a*b
            n=c*c*d
            if m==n: e=1
            if e==-1:
                St=r"\frac{\sqrt{%s}}{  \sqrt{%s} - (\sqrt{%s})} "% (l,m,n)
            else:
                St=r"\frac{\sqrt{%s}}{  \sqrt{%s} + (\sqrt{%s})} "% (l,m,n)
            Val=sp.sqrt(l) /( sp.sqrt(m) +e* sp.sqrt(n))

            
        else:
            a=random.choice(range(2,18))
            b=random.choice(b_li)
            m=a*a*b
            St=r"\sqrt{%s}"% m
            Val=sp.sqrt(m)

        TE={}
        TE["St"]=St
        TE["Val"]=Val        
        TE["Ans"]="-"                
        TE["OK"]="-"      
        TE["Tips"]="sqrt(2) 可寫 J(2 )"
        NTE.append(TE)    
    return NTE
if __name__=="__main__":
    NTE=Get_PF201_Expr(4)
    displayTable(NTE)    
```

```python
if __name__=="__main__":
    x,y,z=sp.symbols('x,y,z')
    NTE=Get_PF201_Expr(4)
    for i,TE in enumerate(NTE):
        St=TE["St"]
        Val=TE["Val"]
        print(f"第{i}題:")        
        display(Latex(f" $$ {St} $$ " ))
        display(Latex(f"答案錯誤, 標準答案: $ {Val} $"))
        print(TE["Tips"])
        TE["Ans"]=input("請作答:")  
        if Put_PF201_Expr(TE):
            print("OK")
        else:
            print("Wrong")
    displayTable(NTE)    
```
